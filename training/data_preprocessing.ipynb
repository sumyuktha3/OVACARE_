```python
import pandas as pd 
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv(r'C:\userspace\daya\python\samu_proj\PCOS_data.csv')
```

```python
df.head(2)
```

```python
copy_df = df.copy()
```

```python

for col in df.columns:
    unique_values = df[col].unique()
    print(f"Unique values in {col}: {unique_values}")
```

```python
# Remove extra spaces from column names in the list
columns_to_drop = ['Sl. No', 'Patient File No.', 'Blood Group', 'Marraige Status (Yrs)', 
                   'I   beta-HCG(mIU/mL)', 'II    beta-HCG(mIU/mL)', 'FSH(mIU/mL)', 
                   'LH(mIU/mL)', 'FSH/LH', 'TSH (mIU/L)', 'AMH(ng/mL)', 'PRL(ng/mL)', 
                   'Vit D3 (ng/mL)', 'PRG(ng/mL)', 'RBS(mg/dl)', 'BP _Systolic (mmHg)', 
                   'BP _Diastolic (mmHg)', 'Follicle No. (L)', 'Follicle No. (R)', 
                   'Avg. F size (L) (mm)', 'Avg. F size (R) (mm)', 'Endometrium (mm)']

# Remove extra spaces from column names in the DataFrame
copy_df.columns = copy_df.columns.str.strip()

# Drop the specified columns
copy_df.drop(columns=columns_to_drop, inplace=True)
```

```python
copy_df.head(3)
```

```python
print("Missing values:\n", copy_df.isnull().sum())
```

```python


# Fill missing values with the most frequent value
imputer = SimpleImputer(strategy='most_frequent')
copy_df['Fast food (Y/N)'] = imputer.fit_transform(copy_df[['Fast food (Y/N)']])
```
