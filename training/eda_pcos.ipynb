```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = r'C:\userspace\daya\python\samu_proj\PCOS_data.csv'
data = pd.read_csv(file_path)
```

```python
data.head(3)
```

```python
print("\nSummary statistics of numerical columns:")
print(data.describe())
```

```python
# Check for missing values
print("\nMissing values:")
print(data.isnull().sum())
```

```python
# Distribution of target variable
plt.figure(figsize=(8, 6))
sns.countplot(x='PCOS (Y/N)', data=data)
plt.title('Distribution of PCOS (Y/N)')
plt.xlabel('PCOS (Y/N)')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of Age
plt.figure(figsize=(8, 6))
sns.histplot(data[' Age (yrs)'], bins=20, kde=True)  # Use the correct column name with leading space
plt.title('Distribution of Age')
plt.xlabel('Age (years)')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation heatmap with adjusted parameters
plt.figure(figsize=(14, 12))  # Increase figure size
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=1.5)  # Adjust cmap and linewidths
plt.title('Correlation Heatmap')
plt.xticks(fontsize=10)  # Adjust x-axis label font size
plt.yticks(fontsize=10)  # Adjust y-axis label font size
plt.show() 
```

```python
# Boxplot of BMI by PCOS status
plt.figure(figsize=(8, 6))
sns.boxplot(x='PCOS (Y/N)', y='BMI', data=data)
plt.title('BMI by PCOS (Y/N)')
plt.xlabel('PCOS (Y/N)')
plt.ylabel('BMI')
plt.show()
```

```python
# Box plot of numerical variables grouped by PCOS status
numerical_columns = ['Age (yrs)', 'Weight (Kg)', 'Height(Cm)', 'BMI', 'Pulse rate(bpm)', 'RR (breaths/min)', 'Hb(g/dl)', 'Cycle length(days)', 'Marraige Status (Yrs)', 'No. of abortions', 'I_beta_HCG_mIU_per_mL', 'II beta-HCG(mIU/mL)', 'FSH(mIU/mL)', 'LH(mIU/mL)', 'FSH/LH', 'Hip(inch)', 'Waist(inch)', 'Waist:Hip Ratio', 'TSH (mIU/L)', 'AMH(ng/mL)', 'PRL(ng/mL)', 'Vit D3 (ng/mL)', 'PRG(ng/mL)', 'RBS(mg/dl)', 'BP _Systolic (mmHg)', 'BP _Diastolic (mmHg)', 'Follicle No. (L)', 'Follicle No. (R)', 'Avg. F size (L) (mm)', 'Avg. F size (R) (mm)', 'Endometrium (mm)']

for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='PCOS (Y/N)', y=data.iloc[:, numerical_columns.index(column)], data=data)
    plt.title(f'Box plot of {column} by PCOS status')
    plt.xlabel('PCOS (Y/N)')
    plt.ylabel(column)
    plt.show()
```

```python
# Bar plot of categorical variables grouped by PCOS status
categorical_columns = ['Blood Group', 'Cycle(R/I)', 'Pregnant(Y/N)', 'Weight gain(Y/N)', 'hair growth(Y/N)', 'Skin darkening (Y/N)', 'Hair loss(Y/N)', 'Pimples(Y/N)', 'Fast food (Y/N)', 'Reg.Exercise(Y/N)']
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=column, hue='PCOS (Y/N)', data=data)
    plt.title(f'Bar plot of {column} by PCOS status')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.legend(title='PCOS (Y/N)')
    plt.show()
```
